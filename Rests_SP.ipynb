# WebScraping

## Get Links

```python
driver = webdriver.Chrome('chromedriver.exe')

# Loop Through - Search Pages that we wanted
lst_urls_sp = []

for i in range(1, 829):
    print('Abrindo Search Pages ' + str(i))
    driver.get('https://www.zomato.com/pt/sao-paulo-sp/restaurantes?page={}'.format(i))
    print('Acessando Webpage OK \n')
    url_rests_jp = driver.find_elements_by_class_name("result-title")

# Loop Through Lists of Web Elements
    for j in url_rests_jp:
        url_sp = j.get_attribute("href")
        lst_urls_sp.append(url_sp)

driver.close()
```

```python
# Colocar lista em um DataFrame
lst_urls_sp = pd.DataFrame(lst_urls_sp, columns=['Website'])
lst_urls_sp
```

```python
# Remover duplicadas
lst_urls_sp = lst_urls_sp.drop_duplicates()
lst_urls_sp
```

```python
# Salvar DF em um arquivo .csv
lst_urls_sp.to_csv('data/sp_links.csv')
```

# Extrair dados e colocar em um DataFrame e nomear colunas

```python
driver = webdriver.Chrome('chromedriver.exe')

rests_sp = []
for url in (lst_urls_sp['Website']):
    print('Abrindo link ' + url)
    driver.get(url)
    try:
        rest_price = driver.find_elements_by_tag_name('p')
        price_temp = []
        for rest_price_anchor in rest_price:
            rest_price_text = rest_price_anchor.text
            if rest_price_text.startswith('R$'):
                price_temp.append(rest_price_text)
        price_temp = price_temp[0]
    except:
        price_temp = None
    try:
        rest_end = driver.find_elements_by_xpath('//*[@id="root"]/div/main/div/section[4]/section/article/section/p')[0].text
    except:
        rest_end = None
    try:
        rest_type = driver.find_element_by_xpath('//*[@id="root"]/div/main/div/section[3]/section/section[1]/section[1]/div').text
    except:
        rest_type = None
    try:
        rest_name = driver.find_element_by_tag_name('h1').text
    except:
        rest_name = None
    try:
        rest_coord = driver.find_element_by_xpath('//*[@id="root"]/div/main/div/section[4]/section/article/section/div[2]/a').get_attribute('href')
    except:
        rest_coord = None
    try:
        rest_bairro = driver.find_element_by_xpath('//*[@id="root"]/div/main/div/section[3]/section/section[1]/section[1]/a').text
    except:
        rest_bairro = None
    
    rests_sp.append([rest_name,price_temp,rest_type,rest_bairro,rest_end,rest_coord,url])
    

driver.close()
```

```python
rests_sp = pd.DataFrame(rests_sp,columns = ['Name','Price','Type','Bairro','Address','Coordenada','Url'])
rests_sp
```
